# VideoAnalyst
## 비디오 로딩 & 포즈 예측 & 저장

```python
data_folders = ['./data/Ours/추가분/FALL', 
                './data/Ours/추가분/WALK', 
                './data/Ours/추가분/SLEEP', 
                './data/Ours/추가분/SIT', 
                './data/Ours/추가분/BEND']
x_data, y_data = [], []
```

### VideoSkelton 이용
- 비디오 스켈레톤 클래스는 비디오 파일로부터 프레임을 추출
- 포즈를 찾아 반환함
- 반환받은 포즈는 x_data, y_data 리스트에 저장

```python
import os
from Skeleton.VideoSkeleton import VideoSkeleton
from mxnet import nd, gpu, cpu

ctx = gpu(1)
vs = VideoSkeleton(ctx=ctx)
bbox_thr = 0.6

for i in range(len(data_folders)) :
    path = data_folders[i]
    onlyfiles = next(os.walk(path))[2]
    
    for video in onlyfiles :
        videopath = os.path.join(path, video)
        print("fetching .. {}".format(videopath))
        coords, _ = vs.predict(videopath, 4, bbox_thr, augument=True)
        ground_truth = nd.array([i]).as_in_context(ctx)
        x_data.extend((coords[0], coords[1]))
        y_data.extend((ground_truth, ground_truth))
        print(coords.shape, ground_truth)
```

### 데이터 저장

```python
with open('./data_pkl/our_data_five.pkl', 'wb') as f :
        pickle.dump(x_data, f)
        pickle.dump(y_data, f)
```
